# Scrapping images from a Singapore government agency monitoring traffic and detecting the cars in the images, trucks are not considered as cars

```python
import pandas as pd 
import numpy as np
from bs4 import BeautifulSoup 
import requests 
from selenium import webdriver
from webdriver_manager.firefox import GeckoDriverManager
import os
from os import path
import cv2
import pickle
from PIL import Image
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

## Firefox needed for the scrapping part, skip this part if you only want to detect the cars from the downladed images

```python
# This function extract the images from a given area of the map, selenium is used to navigate and find 
# the elements in the web page of the area. Fct return url links of the images of that area.

def exctract_images_from_area(area_name):
    
    driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())
    driver.get('https://onemotoring.lta.gov.sg/content/onemotoring/home/driving/traffic_information/traffic-cameras.html')
    search_area = driver.find_element_by_id(area_name)
    driver.execute_script("arguments[0].click();", search_area)
    driver.implicitly_wait(3)
    h1 = driver.find_elements_by_class_name('card')

    links = []
    for elements in h1:
        sub_element = elements.find_elements_by_tag_name('img')
        for i in sub_element:
            links.append(i.get_attribute('src'))
    driver.close()
    driver.quit()
    return links
```

```python
# This function fist finds the different areas in the interactive map and stores them in list_areas, then
# for each area the url links of the images are found. Fct return a dictionary of url links for each area

def extract_images_from_all_areas():
    driver = webdriver.Firefox(executable_path=GeckoDriverManager().install())
    driver.get('https://onemotoring.lta.gov.sg/content/onemotoring/home/driving/traffic_information/traffic-cameras.html')
    
    search_area = driver.find_elements_by_class_name("image-showcase.map")
    list_areas = []
    for elem in search_area[0].find_elements_by_class_name("btn.btn-primary"):
        list_areas.append(elem.get_attribute('id'))
    driver.close()
    driver.quit()
    links = {}
    for area in list_areas:
        links_area = exctract_images_from_area(area)
        links[area]=links_area
    return list_areas, links
```

```python
# Save the dict of url links
list_areas, links = extract_images_from_all_areas()
a_file = open("links.pkl", "wb")
pickle.dump(links, a_file)
a_file.close()
```

```python
# Just in case, to avoid re-running the scrapping functions
a_file = open("links.pkl", "rb")
links = pickle.load(a_file)
```

```python
# We create the directories containing the images

directory = 'area_pictures'
if path.exists(directory) == False:
    os.mkdir(directory)
    
for key in links.keys():
    print(key)
    sub_directory = key
    path_images = os.path.join(directory, sub_directory)
    if path.exists(path_images) == False:
        os.mkdir(path_images)
    for i,url in enumerate(links[key]):
        img_data = requests.get(url).content
        with open(path_images+'/image_'+key+'_'+str(i)+'.jpg', 'wb') as handler:
            handler.write(img_data)
```

## In this part we download the model and predict the position of the cars and their numbers

```python
# We will use a pre-trained yolo model to detect the cars
weightsPath = 'darknet/yolov3.weights'
configPath = 'darknet/cfg/yolov3.cfg'
net = cv2.dnn.readNetFromDarknet(configPath, weightsPath)
```

```python
# Label names and color of binding boxes
labelsPath = 'darknet/data/coco.names'
LABELS = open(labelsPath).read().strip().split("\n")
    # initialize a list of colors to represent each possible class label
np.random.seed(42)
COLORS = np.random.randint(0, 255, size=(len(LABELS), 3),dtype="uint8")
```

```python
# Fct to detect the cars and their number in each image, returns image as numpy array and the number of cars in 
# that image 
# Credit goes to https://opencv-tutorial.readthedocs.io/en/latest/yolo/yolo.html, slightly modified to only take 
# into account the cars
def inference(image, net=net):
    (H, W) = image.shape[:2]

    ln = net.getLayerNames()
    ln = [ln[i - 1] for i in net.getUnconnectedOutLayers()]
    blob = cv2.dnn.blobFromImage(image, 1 / 255.0, (416, 416),swapRB=True, crop=False)
    net.setInput(blob)

    layerOutputs = net.forward(ln)


    boxes = []
    confidences = []
    classIDs = []

    for output in layerOutputs:
        # loop over each of the detections
        for detection in output:
            # extract the class ID and confidence (i.e., probability) of
            # the current object detection
            scores = detection[5:]
            classID = np.argmax(scores)
            confidence = scores[classID]
            # filter out weak predictions by ensuring the detected
            # probability is greater than the minimum probability
            if confidence > 0.5 and classID==2:
                # scale the bounding box coordinates back relative to the
                # size of the image, keeping in mind that YOLO actually
                # returns the center (x, y)-coordinates of the bounding
                # box followed by the boxes' width and height
                box = detection[0:4] * np.array([W, H, W, H])
                (centerX, centerY, width, height) = box.astype("int")
                # use the center (x, y)-coordinates to derive the top and
                # and left corner of the bounding box
                x = int(centerX - (width / 2))
                y = int(centerY - (height / 2))
                # update our list of bounding box coordinates, confidences,
                # and class IDs
                boxes.append([x, y, int(width), int(height)])
                confidences.append(float(confidence))
                classIDs.append(classID)
                

    idxs = cv2.dnn.NMSBoxes(boxes, confidences, 0.5,0.3)
    
    if len(idxs) > 0:
        
        # loop over the indexes we are keeping
        for i in idxs.flatten():
            # extract the bounding box coordinates
            (x, y) = (boxes[i][0], boxes[i][1])
            (w, h) = (boxes[i][2], boxes[i][3])
            # draw a bounding box rectangle and label on the image
            color = [int(c) for c in COLORS[classIDs[i]]]
            cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
            text = "{}: {:.4f}".format(LABELS[classIDs[i]], confidences[i])
            cv2.putText(image, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX,
                0.5, color, 2)
            
    return image, len(idxs)
```

```python
# Run this to predict on one particular image

image = cv2.imread('area_pictures/aye/image_aye_9.jpg')
im, number_of_cars = inference(image, net)
print('number of cars in image is ',number_of_cars)
figure(figsize=(20, 15), dpi=80)
plt.imshow(im, interpolation='nearest')
plt.show()
```

```python
# RUn this to find the cars in all of the images
directory = 'output_area_pictures'
if path.exists("output_area_pictures") == False:
    os.mkdir(directory)
    
for key in links.keys():
    sub_directory = key
    path_images = os.path.join(directory, sub_directory)
    if path.exists(path_images) == False:
        os.mkdir(path_images)
        
    for i,filename in enumerate(os.listdir('area_pictures/'+key+'/')):
        image_path = os.path.join('area_pictures/'+key+'/', filename)
        image = cv2.imread(image_path)
        car_inf, _ = inference(image, net)
        prediction = Image.fromarray(car_inf)
        prediction.save(os.path.join(path_images,filename))
```
